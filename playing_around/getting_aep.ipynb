```python
from Farm_Evaluator_Vec import find_aep
import pandas as pd
import numpy as np
```

```python
import os 
from glob import glob
```

```python
WIND_DATA_270 = 'wind_data/wind_270.csv'
WIND_DATA_225 = 'wind_data/wind_225.csv'
WIND_DATA_224 = 'wind_data/wind_224.csv'
WIND_DATA_180 = 'wind_data/wind_180.csv'
WIND_DATA_FULL = 'wind_data/full_wind_data.csv'

POWER_CURVE = 'power_curve.csv'
```

```python
files = glob('turbine_data/*csv')
```

# wind data = 270

```python
wind_data = pd.read_csv(WIND_DATA_270)
wind_data.head()
```

```python
wind_data.describe()
```

```python
aep_270 = {}
for file in files:
    name = file.split('/')[-1]
    aep = find_aep(file,POWER_CURVE,WIND_DATA_270)
    aep_270[name] = aep
    print(aep)
```

```python
df = pd.DataFrame(aep_270,index = [1]).T
df.to_csv('scores/scores_wind_270')
```

```python
df.sort_values(by = 1,ascending = False)
```
