```
from tqdm import tqdm

NUM_TURBINES = 20

x = np.arange(50,3950,10)
y = np.arange(50,3950,10)
points = [(a,b) for a in x for b in y]
total_points = []

for _ in tqdm(range(NUM_TURBINES)):
    random_point_idx =  np.random.choice(len(points))
    random_point = points[random_point_idx]
    total_points.append(random_point)
    points_in_region = []
    for point in points:
      if point[0] in range(random_point[0] - 400,random_point[0] + 400 + 1,10) and point[1] in range(random_point[1] - 400,random_point[1] + 400 + 1):
        points_in_region.append(point)
    points = list(set(points) - set(points_in_region))

plt.figure(figsize = (10,8))
plt.scatter(np.array(points)[:,0],np.array(points)[:,1])
plt.scatter(np.array(total_points)[:,0],np.array(total_points)[:,1],c = 'orange',s = 100)
plt.xlim(0,4000)
plt.ylim(0,4000)

```
